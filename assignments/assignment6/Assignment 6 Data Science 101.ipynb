# Chi Square Test

```python
import pandas as pd
from scipy.stats import chi2_contingency

#I used Chatgpt to help me with this test 
# Define column names (based on the dataset documentation)
columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 
    'marital-status', 'occupation', 'relationship', 'race', 
    'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 
    'native-country', 'income'
]

# Load the dataset from the URL
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
df = pd.read_csv(url, names=columns, sep=',\s', engine='python')

# Display the first few rows to ensure its loaded correctly
df.head()


# Generate a contingency table for 'education' and 'income'
contingency_table = pd.crosstab(df['education'], df['income'])

# Display the contingency table
print(contingency_table)

```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-Square Statistic = {chi2}")
print(f"P-Value = {p}")
print(f"Degrees of Freedom = {dof}")
print("Expected Frequencies:")
print(expected)
```

```python
#Here we accept the null hypothesis because p-value=0.There is statistical significance between
#education and income. The chi-square is large value and there is big difference between observed and
#expect frequencies and satistically significant. 
```

```python
from scipy.stats import f_oneway

# Separate the data into different education levels
hs_hours = df[df['education'] == 'HS-grad']['hours-per-week']
bachelors_hours = df[df['education'] == 'Bachelors']['hours-per-week']
masters_hours = df[df['education'] == 'Masters']['hours-per-week']

```

```python
f_stat, p_value = f_oneway(hs_hours, bachelors_hours, masters_hours)

# Print the result
print(f"F-Statistic = {f_stat}")
print(f"P-Value = {p_value}")
```

```python
#The F statistic is large meaning that meaning there is a big variance in number of hours worked across having a highschool education level, bachelors education 
#level and a masters education level. The p-value is above 0.05 meaning education level has a significant effect on the hours worked. 
```
